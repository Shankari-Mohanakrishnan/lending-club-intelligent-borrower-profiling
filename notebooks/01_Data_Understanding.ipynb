## Missing Value Analysis

Missing values are calculated for each column.

Columns with very high missing percentages may need to be removed because they provide little usable information for modeling.

This step helps identify:
- completely empty columns
- sparse hardship or settlement columns
- columns requiring imputation

```python
missing_summary= df.isnull().sum().to_frame('missing_count')\
    .assign(missing_percent=lambda x: (x['missing_count']/len(df))*100)\
    .sort_values('missing_percent',ascending=False)

missing_summary.head(30)
```

## Numerical Feature Summary

```python
# Basic Numeric Summary
df.describe().T.head(30)
```

## Categorical and Numerical Column Split

Columns are separated into categorical and numerical groups.

This separation is useful because each type requires different preprocessing:

- categorical columns may need encoding
- numerical columns may need scaling, imputation, or outlier handling

```python
# Basic Object/categorical summary
cat_cols = df.select_dtypes(include='object').columns.tolist()
num_cols = df.select_dtypes(exclude='object').columns.tolist()

print(f'Categorical Columns:', len(cat_cols))
print(f'Numeric Columns:', len(num_cols))
```

## Column-Level Data Quality Overview

A column overview table is created to summarize:

- column name
- data type
- missing count
- missing percentage
- number of unique values

This provides a high-level data quality report and helps decide which columns to keep, drop, or transform in later notebooks.

```python
columns_overview = pd.DataFrame({
    'column_name': df.columns,
    'dtype': df.dtypes.astype(str).values,
    'missing_count': df.isnull().sum().values,
    'missing_percent': (df.isnull().sum().values/len(df))*100,
    'unique_values': df.nunique(dropna=False).values
    
})

columns_overview=columns_overview.sort_values('missing_percent',ascending=False)
columns_overview.head(30)
```
