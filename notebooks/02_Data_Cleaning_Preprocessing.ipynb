```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns',100)
```

```python
data['loan_status'].unique()
```

```python
good_status = ['Does not meet the credit policy. Status:Fully Paid','Fully Paid']
bad_status = ['In Grace Period', 'Does not meet the credit policy. Status:Charged Off','Late (31-120 days)','Late (16-30 days)','Charged Off', 'Default']
df = data[data['loan_status'].isin(good_status + bad_status)].copy()
print(f'Shape before keeping  only good & bad statuses: {data.shape}')
print(f'Shape after keeping  only good & bad statuses: {df.shape}')
df.head()
```

```python
df['loan_status'].unique()
```

Current staus is removed as we are considering only good (completed pay back) and bad (not paying regularly).

```python
df['default_flag']=np.where(df['loan_status'].isin(bad_status),1,0)
df.head()
```

```python
df['default_flag'].value_counts()
```

```python
df['default_flag'].value_counts(normalize=True)*100
```

22% of data is in bad status (charge off, default, not paid) and 77% in good status (paid back). 

## Dropping columns having more than 50% missing columns 

```python
missing_percent = df.isnull().mean()*100
cols_to_drop = missing_percent[missing_percent>50].index.to_list()
print(f'Number of columns that have more than 50% data missing {len(cols_to_drop)}')
print("\n\nThe columns are:")
print(cols_to_drop)
df = df.drop(columns=cols_to_drop)
print(f'\n\nThe shape after dropping the coumns: {df.shape}')
```
