# Feature Engineering for Credit Risk Modeling

## Objective

This notebook transforms the cleaned Lending Club dataset into a machine learning-ready dataset by engineering predictive features that better represent borrower behavior, repayment capacity, financial health, and credit risk.

The feature engineering process includes:

- Converting categorical values into numerical features
- Creating date-derived variables
- Engineering financial ratio features
- Building credit utilization indicators
- Creating delinquency and bankruptcy flags
- Ordinal encoding of Lending Club grades
- Constructing borrower profile scores
- Capping extreme ratio values
- Evaluating engineered feature relationships with the target variable

The resulting dataset will be used for predictive default modeling.

## Import Required Libraries

Load the required Python libraries for feature engineering and data manipulation.

```python
import pandas as pd 
import numpy as np

pd.set_option('display.max_columns',100)
```

## Load Cleaned Dataset

The cleaned Lending Club dataset produced in the previous preprocessing notebook is loaded for feature engineering.

```python
df = pd.read_pickle('../data/lending_club_clean_base.pkl')
print(df.shape)
df.head()
```

## Convert Loan Term into Numeric Feature

The original loan term is stored as text (for example, "36 months").

A numerical version is created so that machine learning algorithms can directly utilize loan duration as a predictive feature.

```python
df['term'].unique()
```

```python
df['term_months']=df['term'].str.extract('(\d+)').astype(float)
df['term_months'].unique()
```

## Convert Employment Length to Numeric Values

Employment length is converted from categorical text into numerical years.

Longer employment history generally indicates greater income stability and may influence repayment behavior.

```python
Number_of_unique_emp_len=df['emp_length'].nunique()
print(f'Number of unique employment length values {Number_of_unique_emp_len} \n\n')
df['emp_length'].unique()
```

emp_length_map = {
     '10+ years':10,
     '9 years': 9,
     '8 years': 8,
     '7 years': 7,
     '6 years': 6,
     '5 years': 5,
     '4 years': 4,
     '3 years': 3,
     '2 years': 2,
     '1 years': 1,
}

df['emp_length_num']=df['emp_length'].map(emp_length_map)

```python
df.head()
```

## Convert Date Columns

Loan-related date columns are converted into datetime format to enable temporal feature engineering.

```python
date_cols = ["issue_d", "earliest_cr_line", "last_pymnt_d", "last_credit_pull_d"]
df[date_cols].head()
```

```python
for col in date_cols:
    if col in df:
        df[col] = pd.to_datetime(df[col],format='%b-%Y',errors='coerce')

df[date_cols].head(10)
```

## Create Date-Based Features

Additional temporal features are extracted from loan issue dates.

New features include:

- Issue month
- Issue year
- Credit history length

Credit history length represents the number of months between the borrower's earliest credit account and loan issuance.

```python
df['issue_month']=df['issue_d'].dt.month
df['issue_year']=df['issue_d'].dt.year

df['credit_history_months']= ((df['issue_d'].dt.year - df['earliest_cr_line'].dt.year) * 12) - (df['issue_d'].dt.month - df['earliest_cr_line'].dt.month)
#Invalid credit_history_months set to missing
df.loc[df['credit_history_months']<0,'credit_history_months']=np.nan


df[date_cols + ['issue_month', 'issue_year', 'credit_history_months']].head()
```

```python
df[df['credit_history_months']<0]
```

### Data Checks & Preperation before Creating Ratio Columns

#### Validate Denominator Variables

Before creating ratio-based features, denominator variables are inspected for missing, zero, or invalid values.

This prevents divide-by-zero errors and improves feature reliability.

```python
print(f"Number of NULLs in annual_inc {df['annual_inc'].isnull().sum()}")
print(f"Number of NULLs in total_acc {df['total_acc'].isnull().sum()}")
print(f"Number of NULLs in funded_amnt {df['funded_amnt'].isnull().sum()}")
```

```python
print(f"Number of negatives in annual_inc {(df['annual_inc']<= 0).sum()}")
print(f"Number of negatives in total_acc {(df['total_acc']<= 0).sum()}")
print(f"Number of negatives in funded_amnt {(df['funded_amnt']<= 0).sum()}")
```

#### Clean Denominator Variables

Invalid denominator values are replaced with missing values to ensure mathematically valid ratio calculations.

```python
df['annual_inc_clean']=df['annual_inc'].where(df['annual_inc']>0, np.nan)
df['total_acc_clean']=df['total_acc'].where(df['total_acc']>0, np.nan)
df['funded_amnt_clean']=df['funded_amnt'].where(df['funded_amnt']>0, np.nan)
```

```python
print(f"Number of invalid in annual_inc_clean {(df['annual_inc_clean']<= 0).sum()}")
print(f"Number of invalid in total_acc_clean {(df['total_acc_clean']<= 0).sum()}")
print(f"Number of invalid in funded_amnt_clean {(df['funded_amnt_clean']<= 0).sum()}")
```

## Engineer Financial Capacity Features

Financial ratio features are created to measure the relationship between loan obligations and borrower income.

These ratios capture repayment capacity and financial leverage, which are important predictors of credit default.

```python
df.columns
```

```python
df.head()
```

```python
df['loan_to_income_ratio'] = df['loan_amnt']/df['annual_inc_clean']
df['funded_to_income_ratio']= df['funded_amnt']/df['annual_inc_clean']
df['revol_bal_to_income_ratio']=df['revol_bal']/df['annual_inc_clean']
df['installment_to_income_ratio'] = (df['installment']*12)/df['annual_inc_clean']

df[['loan_to_income_ratio', 'funded_to_income_ratio', 'revol_bal_to_income_ratio', 'installment_to_income_ratio']].head(5)
```

## Engineer Credit Utilization Features

Additional features are created to measure account usage and overall debt burden.

These variables summarize how heavily borrowers rely on existing credit before receiving a new loan.

```python
df['open_acc_to_total_acc_ratio'] = df['open_acc']/df['total_acc_clean']

df['dti_clean']=df['dti'].where(df['dti']>=0,np.nan)
df['revol_util_clean']=df['revol_util'].where(df['revol_util']>=0,np.nan)

df['debt_burden_score'] = df['dti_clean'] + df['revol_util_clean']

df[['open_acc_to_total_acc_ratio','dti_clean','revol_util_clean','debt_burden_score']].head()
```


## Create Risk Indicator Flags

Binary risk indicators are created from borrower credit history.

These features capture the presence of:

- Previous delinquencies
- Public records
- Bankruptcies
- High revolving utilization
- High debt-to-income ratio

These variables provide simple yet informative risk signals.

```python
df['has_delinquency'] = np.where(df['delinq_2yrs']>0,1,0)
df['has_public_record'] = np.where(df['pub_rec']>0,1,0)
df['has_bankruptcy'] = np.where(df['pub_rec_bankruptcies']>0,1,0)

df['high_revol_util_flag'] = np.where(df['revol_util_clean']>75,1,0)
df['high_dti_flag'] = np.where(df['dti_clean']>35,1,0)

df[['has_delinquency', 'has_public_record', 'has_bankruptcy', 'high_revol_util_flag', 'high_dti_flag']].head()
```

## Ordinal Encode Lending Club Grades

Lending Club grades naturally represent increasing credit risk.

The alphabetical grades are converted into ordered numerical values while preserving their ranking.

```python
df['grade'].unique()
```

```python
grade_map = {
    'A':1,
    'B':2,
    'C':3,
    'D':4,
    'E':5,
    'F':6,
    'G':7
}
```

```python
df['grade_num'] = df['grade'].map(grade_map)
df[['grade', 'grade_num']]. drop_duplicates().sort_values('grade_num')
```

## Ordinal Encode Loan Sub-Grades

Sub-grades provide finer borrower risk granularity than the primary loan grades.

Each sub-grade is mapped into an ordered numerical scale ranging from A1 to G5.

```python
sorted(df['sub_grade'].unique())
```

```python
sub_grade_order = {
    grade: index+1
    for index, grade in enumerate(sorted(df['sub_grade'].unique()))
}

sub_grade_order
```

```python
df['sub_grade_number'] = df['sub_grade'].map(sub_grade_order)

df[['sub_grade','sub_grade_number']].drop_duplicates().sort_values('sub_grade_number')
```

## Create Borrower Percentile Rankings

Several important borrower characteristics are converted into percentile rankings.

Percentile features standardize variables measured on different scales and provide relative comparisons across the borrower population.

```python
df['income_rank']=df['annual_inc_clean'].rank(pct=True)
df["loan_to_income_rank"] = df["loan_to_income_ratio"].rank(pct=True)
df["installment_to_income_rank"] = df["installment_to_income_ratio"].rank(pct=True)
df["credit_history_rank"] = df["credit_history_months"].rank(pct=True)
df["emp_length_rank"] = df["emp_length_num"].rank(pct=True)
df["dti_rank"] = df["dti_clean"].rank(pct=True)
df["revol_util_rank"] = df["revol_util_clean"].rank(pct=True)
df["grade_risk_rank"] = df["grade_num"].rank(pct=True)
df["sub_grade_risk_rank"] = df["sub_grade_num"].rank(pct=True)
df["delinq_rank"] = df["delinq_2yrs"].rank(pct=True)
df["pub_rec_rank"] = df["pub_rec"].rank(pct=True)


df[['income_rank', "loan_to_income_rank", "installment_to_income_rank", "credit_history_rank", "emp_length_rank", "dti_rank",
    "revol_util_rank", "grade_risk_rank", "sub_grade_risk_rank", "delinq_rank", "pub_rec_rank"]].head()
```

## Build Composite Borrower Profile Scores

Multiple engineered variables are combined into higher-level borrower profile scores.

The composite scores summarize:

- Repayment capacity
- Creditworthiness
- Financial stability

These higher-level representations may improve predictive performance while increasing business interpretability.

```python
df['repayment_capacity_score'] = df['income_rank'] - df['loan_to_income_rank'] - df['installment_to_income_rank']
df['creditworthiness_score'] = df['grade_risk_rank'] + df['sub_grade_risk_rank'] + df['delinq_rank'] + df['pub_rec_rank']
df['financial_stability_score'] = df['credit_history_rank'] + df['emp_length_rank'] - df['dti_rank'] - df['revol_util_rank']

df[['repayment_capacity_score', 'creditworthiness_score', 'financial_stability_Score']].head()
```

## Cap Extreme Ratio Values

Extreme values in ratio features are capped using the 1st and 99th percentiles.

This reduces the influence of outliers while preserving the overall distribution of the data.

```python
ratio_cols = ['loan_to_income_ratio', 'funded_to_income_ratio', 'revol_bal_to_income_ratio', 'installment_to_income_ratio', 
              'open_acc_to_total_acc_ratio', 'debt_burden_score']

for col in ratio_cols:
    lower = df[col].quantile(0.01)
    upper = df[col].quantile(0.99)
    df[col+'_capped'] = df[col].clip(lower=lower, upper=upper)

df[['loan_to_income_ratio_capped', 'funded_to_income_ratio_capped', 'revol_bal_to_income_ratio_capped', 
     'installment_to_income_ratio_capped', 'open_acc_to_total_acc_ratio_capped', 'debt_burden_score_capped']].head()
```

## Review Engineered Features

A consolidated list of newly engineered variables is created to simplify downstream modeling and analysis.

```python
engineered_cols = [
    "term_months",
    "emp_length_num",
    "issue_year",
    "issue_month",
    "credit_history_months",
    "loan_to_income_ratio",
    "installment_to_income_ratio",
    "funded_to_income_ratio",
    "revol_bal_to_income_ratio",
    "open_acc_to_total_acc_ratio",
    "debt_burden_score",
    "loan_to_income_ratio_capped",
    "installment_to_income_ratio_capped",
    "funded_to_income_ratio_capped",
    "revol_bal_to_income_ratio_capped",
    "open_acc_to_total_acc_ratio_capped",
    "debt_burden_score_capped",
    "has_delinquency",
    "has_public_record",
    "has_bankruptcy",
    "high_revol_util_flag",
    "high_dti_flag",
    "grade_num",
    "sub_grade_num",
    "repayment_capacity_score",
    "creditworthiness_score",
    "financial_stability_score"
]

df[engineered_cols+['default_flag']].head()
```

## Evaluate Missing Values in Engineered Features

The engineered features are reviewed for missing values to determine whether additional imputation will be required during model training.

```python
engineered_missing = df[engineered_cols].isnull().sum().to_frame('missing_count')\
    .assign(missing_percent=lambda x: (x['missing_count']/len(df))*100)\
    .sort_values('missing_percent', ascending=False)

engineered_missing[engineered_missing['missing_percent']>0]
```

## Relationship Between Engineered Features and Default

The correlation between engineered variables and the default target is examined.

This provides an initial indication of which engineered features are most informative for predicting loan default.

```python
feature_corr = df[engineered_cols+['default_flag']].corr(numeric_only=True)['default_flag'].sort_values(ascending=False)

feature_corr
```

## Analyze Default Rates by Risk Indicators

Default rates are compared across key engineered risk indicators.

This validates whether the engineered features capture meaningful differences in borrower default behavior.

```python
flag_cols = [
    "has_delinquency",
    "has_public_record",
    "has_bankruptcy",
    "high_revol_util_flag",
    "high_dti_flag",
    "term_months"
]

for col in flag_cols:
    display(df.groupby(col)['default_flag'].agg(['count','mean']).rename(columns={'count':'default_count','mean':'default_rate'}).sort_index())
```
